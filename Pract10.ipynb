```python
import seaborn as sb
import matplotlib.pyplot as plt
import pandas as pd
```

```python
ds = sb.load_dataset('iris')

ds.head()
```

```python
ds.info()
```

```python
print(ds.columns)
```

```python
print(ds.dtypes)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(16,9))

# Sepal Length
sb.histplot(ds['sepal_length'], ax=axes[0,0])

# Sepal Width
sb.histplot(ds['sepal_width'], ax=axes[0,1])

# Petal Length
sb.histplot(ds['petal_length'], ax=axes[1,0])

# Petal Width
sb.histplot(ds['petal_width'], ax=axes[1,1])

plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(16,9))

# Petal Length
sb.boxplot(
    x='species',
    y='petal_length',
    data=ds,
    ax=axes[0,0]
)

# Petal Width
sb.boxplot(
    x='species',
    y='petal_width',
    data=ds,
    ax=axes[0,1]
)

# Sepal Length
sb.boxplot(
    x='species',
    y='sepal_length',
    data=ds,
    ax=axes[1,0]
)

# Sepal Width
sb.boxplot(
    x='species',
    y='sepal_width',
    data=ds,
    ax=axes[1,1]
)

plt.show()
```
