```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
df = sb.load_dataset("titanic")

df.head()
```

```python
df.info()
```

```python
print("Missing Values : ")
df.isnull().sum()
```

```python
df["age"].fillna(df["age"].median(), inplace = True)
```

```python
df.isnull().sum()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(8,4))

fig.suptitle("Histogram 1-Variables (Age & Fare)")

sb.histplot(
    data = df,
    x = "age",
    ax = axes[0]
)

sb.histplot(
    data=df,
    x = "fare",
    ax = axes[1]
)

plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(16,9))

fig.suptitle('Histogram 2-Variables')

# Age vs Survived
sb.histplot(
    data=df,
    x='age',
    hue='survived',
    multiple='dodge',
    ax=axes[0][0]
)

# Fare vs Survived
sb.histplot(
    data=df,
    x='fare',
    hue='survived',
    multiple='dodge',
    ax=axes[0][1]
)

# Age vs Gender
sb.histplot(
    data=df,
    x='age',
    hue='sex',
    multiple='dodge',
    ax=axes[1][0]
)

# Fare vs Gender
sb.histplot(
    data=df,
    x='fare',
    hue='sex',
    multiple='dodge',
    ax=axes[1][1]
)

plt.show()
```
